```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
import os

def soup(url):
    thepage = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(thepage, "html.parser")
    thepage.close()
    return soupdata

edatas = ""
edata1 = ""
data = ""

surl = "https://ev-database.org/#sort:path~type~order=.rank~number~desc|range-slider-range:prev~next=0~1200|range-slider-acceleration:prev~next=2~23|range-slider-topspeed:prev~next=110~450|range-slider-battery:prev~next=10~200|range-slider-towweight:prev~next=0~2500|range-slider-fastcharge:prev~next=0~1500|paging:currentPage=0|paging:number=9"
soup1 = soup(surl)
#print(soup1)

count = soup1.findAll("span", {"class": "model"})
print(len(count))
```

```python
for record in soup1.findAll("div",{"class":"list"}):
    for record2 in record.findAll("div",{"class": "data-wrapper"}):
        for record3 in record2.findAll("a",{"class": "title"}):
            model = record3.text.replace(',','..')
        for record4 in record2.findAll("span",{"class": "efficiency"}):
            eff = record4.text.replace(',','..')
        for record4 in record2.findAll("span",{"class": "topspeed"}):
            ts = record4.text.replace(',','..')
        for record4 in record2.findAll("span",{"class": "acceleration"}):
            acc = record4.text.replace(',','..')
        for record4 in record2.findAll("span",{"class": "erange_real"}):
            ran = record4.text.replace(',','..')
        for record4 in record2.findAll("span",{"class": "fastcharge_speed_print"}):
            fc = record4.text.replace(',','..')
        for record4 in record2.findAll("span",{"class": "country_uk"}):
            puk = record4.text.replace(',','')
        for record4 in record2.findAll("sapn",{"title":"Number of seats"}):
            seat = record4.text.replace(',','..')
            data = model+","+eff+","+ts+","+acc+","+ran+","+fc+","+puk+','+seat
        edatas = edatas +"\n"+ data
print(edatas)


```

```python
header = "Vehicle_name,Efficiency,Top_speed,Acceleration,Range,FastChargeSprint,PriceinUK,NoOfSeats"
file = open(os.path.expanduser("out.csv"), "wb")
file.write(bytes(header, encoding="ascii", errors="ignore"))
file.write(bytes(edatas, encoding="ascii", errors="ignore"))
```
